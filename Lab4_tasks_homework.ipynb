```
#Task 2
import string
fin=open("mybook.txt")
punct=string.punctuation+string.whitespace
lst=[]
diction={}
for line in fin:
  word=line.split()
  for word1 in word:
   word1=word1.strip(punct)
   word1=word1.lower()
   if word1=="http://pglaf.org/donate":
    word1="http://pglaf.org/donate".replace("http://pglaf.org/donate","pglaf.org")     
   lst.append(word1)
print(len(lst))
for i in lst:
 if i not in diction:
    diction[i]=1
 else:
    diction[i]=diction[i]+1
#print(diction)
finalresult=sorted(((value,key) for key,value in diction.items()),reverse=True)
print(finalresult)
#final_res=list(finalresult)
#print(final_res)
for freq,word in finalresult[:20]:
 print(word, freq,sep="\t")
```
